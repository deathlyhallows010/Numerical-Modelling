```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D 
%matplotlib inline
```

```python
x = np.arange(-5,5,.1);
y = np.arange(-5,5,.1);

Rth = np.zeros((len(x),len(y)));
Rcal = np.zeros((len(x),len(y)));
```

```python
fth = lambda x,ux,sx,y,uy,sy : (1/(2*np.pi*sx*sy)) * np.exp(-(((x-ux)**2)/(2*sx**2) + ((y-uy)**2)/(2*sy**2)))
fcal = lambda x,ux,sx,y,uy,sy : np.exp(-(((x-ux)**2)/(2*sx**2) + ((y-uy)**2)/(2*sy**2)))
```

```python
for i in range(len(x)):
    for j in range(len(y)):
        Rth[i][j] = fth(x[i],0,1,y[j],0,1)
        Rcal[i][j] = fcal(x[i],0,1,y[j],0,1)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
 
X, Y = np.meshgrid(x, y)
# Plot the surface.
surf = ax.plot_surface(X,Y,Rth,cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

plt.title('Theoretical f(x,y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
