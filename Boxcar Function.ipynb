```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def boxcar(x,n):
    p = np.pi/2
    for i in range(1,n,2):
        p += (2/i)*np.sin(i*x)
    return p
```

```python
x = np.arange(-2*np.pi,2*np.pi,0.01)
```

```python
plt.plot(x,boxcar(x,100000))
plt.show()
```

```python
x = np.arange(-np.pi,np.pi,0.01)


fig, ax = plt.subplots()
ax.plot(x,boxcar(x,100000))

n = [10,30,150,250,500]

legend_lst = ['Boxcar function']

for i in n:
    ax.plot(x,boxcar(x,i))

for i in n:
    legend_lst.append(f'n - {i} terms')
ax.legend(legend_lst, loc=2)
ax.set_xlim([-np.pi,np.pi])

plt.show()
```
