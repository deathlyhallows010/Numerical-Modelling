```python
import sympy as sy
import numpy as np
from sympy.functions import sin,cos
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pi = np.pi
```

```python
# Define the variable and the function to approximate
x = sy.Symbol('x')
f_cos = cos(x)
f_sin = sin(x)
```

```python
# Factorial function
def factorial(n):
    if n <= 0:
        return 1
    else:
        return n*factorial(n-1)

# Taylor approximation at x0 of the function 'function'
def taylor(function,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (function.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p
```

```python
x_lims = [-5,5]
x1 = np.linspace(-5,5,800)
y1 = []

func = taylor(f_cos,pi/3,3)
print('Taylor expansion at n='+str(3),func)
for k in x1:
    y1.append(func.subs(x,k))
plt.plot(x1,y1,label='order '+str(3))
y1 = []
# Plot the function to approximate (sine, in this case)
plt.plot(x1,np.cos(x1),label='cos of x')
plt.scatter(pi/3,cos(pi/3))
plt.xlim(x_lims)
plt.ylim([-5,5])
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Taylor series approximation of Cos')
plt.show()
```

```python
x_lims = [-5,5]
x1 = np.linspace(-5,5,800)
y1 = []

func = taylor(f_sin,pi/3,3)
print('Taylor expansion at n='+str(3),func)
for k in x1:
    y1.append(func.subs(x,k))
plt.plot(x1,y1,label='order '+str(3))
y1 = []
# Plot the function to approximate (sine, in this case)
plt.plot(x1,np.sin(x1),label='sin of x')
plt.scatter(np.pi/3,np.sin(pi/3))
plt.xlim(x_lims)
plt.ylim([-5,5])
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Taylor series approximation of Sin')
plt.show()
```

```python
f = cos(x)
# Plot results
def plot():
    x_lims = [-5,5]
    x1 = np.linspace(x_lims[0],x_lims[1],800)
    y1 = []
    # Approximate up until 10 starting from 1 and using steps of 2
    for j in range(1,10,2):
        func = taylor(f,pi/3,j)
        print('Taylor expansion at n='+str(j),func)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1,y1,label='order '+str(j))
        y1 = []
    # Plot the function to approximate (sine, in this case)
    plt.plot(x1,np.sin(x1),label='sin of x')
    plt.xlim(x_lims)
    plt.ylim([-5,5])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title('Taylor series approximation')
    plt.show()

plot()
```
