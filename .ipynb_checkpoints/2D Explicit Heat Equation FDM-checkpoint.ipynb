```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Physical parameters

```python
L = 150e3 # Width of lithosphere [m]
H = 100e3 # Height of lithosphere [m]
Tbot = 1300 # Temperature of bottom lithosphere [C]
Tsurf = 0 # Temperature of country rock [C]
Tplume = 2500 # Temperature of plume [C]
kappa = 1e-6 # Thermal diffusivity of rock [m2/s]
Wplume = 25e3 # Width of plume [m]
day = 3600*24 # seconds per day
year = 365.25*day # seconds per year
```

### Numerical parameters

```python
nx = 101 # gridpoints in x-direction
nz = 51 # gridpoints in z-direction
nt = 1000 # Number of timesteps to compute
dx = L/(nx-1) # Spacing of grid in x-direction
dz = H/(nz-1) # Spacing of grid in z-direction
x2d,z2d = np.meshgrid(np.arange(-L/2,L/2+dx,dx) , np.arange(-H,0+dz,dz) ) # create grid
```

```python
# Compute stable timestep
dt = min([dx,dz])**2/kappa/4;
# Setup initial linear temperature profile
T = abs(z2d/H)*Tbot;
# Imping plume beneath lithosphere
ind = (abs(x2d[0,:]) <= Wplume/2)
T[0,ind] = Tplume
```

```python
%%time

time = 0

for n in range(nt):
    # Compute new temperature
    Tnew = np.zeros((nz,nx))
    sx = kappa*dt/dx**2
    sz = kappa*dt/dz**2
    for j in range(1,nx-1):
        for i in range(1,nz-1):
            Tnew[i,j] = sx*T[i,j+1]+(1-2*sx-2*sz)*T[i,j]+sx*T[i,j-1]+sz*T[i+1,j]+sz*T[i-1,j]

    # Set boundary conditions
    Tnew[0,:] = T[0 ,:]
    Tnew[nz-1,:] = T[nz-1 ,:]
    for i in range(1,nz-1):
        Tnew[i,0] = T[i,0]
        Tnew[i,nx-1] = T[i,nx-1]
        
    T = Tnew
    time = time+dt

    
levels = np.arange(0,2400+100,100)

cp = plt.contourf(x2d/1e3,z2d/1e3,Tnew,levels=levels)
plt.colorbar(cp)
cp.ax.set_ylabel('Temperature [^oC]', rotation=270)


plt.xlabel('x [km]')
plt.ylabel('z [km]')
plt.title('Temperature evolution after '+str(time/year/1e6)+' Myrs')
plt.show()
```
