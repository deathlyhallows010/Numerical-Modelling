```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func_cos(x, n):
    cos_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i)
        denom = math.factorial(2*i)
        cos_approx += ( coef ) * ( (num)/(denom) )
    
    return cos_approx

def func_sin(x, n):
    sin_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i+1)
        denom = math.factorial(2*i+1)
        sin_approx += ( coef ) * ( (num)/(denom) )
    
    return sin_approx
```

```python
angle_rad = (math.radians(60))
out = func_cos(angle_rad,3)
print(out)
```

```python
angle_rad = (math.radians(60))
out = func_sin(angle_rad,3)
print(out)
```

```python
import numpy as np

def f(x):
    return x ** 2
#For instance f(2) = 4

def derive(function, value):
    h = 0.00000000001
    top = function(value + h) - function(value)
    bottom = h
    slope = top / bottom
    # Returns the slope to the third decimal
    return float("%.3f" % slope)
```

```python
derive(f, 2)
```
