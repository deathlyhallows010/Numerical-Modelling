```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
points = [[random.randint(0, 10) for i in range(2)] for j in range(10)]
print(points)
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d
vor = Voronoi(points)

types = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
voronoi_plot_2d(vor)
for i,type in enumerate(types):
    x = points[i][0]
    y = points[i][1]
    plt.text(x+0.3, y+0.3, type, fontsize=9)
plt.xlim([0,10])
plt.ylim([0,10])
plt.show()
```

```python
vor.vertices
```

```python
points_1 = [[random.randint(1, 9) for i in range(2)] for j in range(5)]
types_1 = ['1','2','3','4','5']

voronoi_plot_2d(vor)
for i,type in enumerate(types):
    x = points[i][0]
    y = points[i][1]
    plt.text(x+0.3, y+0.3, type, fontsize=9)
for i,type in enumerate(types_1):
    x = points_1[i][0]
    y = points_1[i][1]
    plt.scatter(x, y, marker='x', color='red')
    plt.text(x+0.3, y+0.3, type, fontsize=9)
    
plt.xlim([0,10])
plt.ylim([0,10])
plt.show()
```

```python
from scipy.spatial import distance

result = []

for i in range(len(points_1)):
    min_dist = distance.euclidean((points_1[i][0],points_1[i][1]), (points[0][0],points[0][1]))
    min_p = 0
    for j in range(1,len(points)):
        if min_dist > distance.euclidean((points_1[i][0],points_1[i][1]), (points[j][0],points[j][1])):
            min_p = j;
            min_dist = distance.euclidean((points_1[i][0],points_1[i][1]), (points[j][0],points[j][1]))
    result.append(types[min_p])
        
```

```python
print('Classes assosiated with points :\n')
for i in range(len(points_1)):
    print('(',points_1[i][0],',',points_1[i][1],')', ' -->', result[i])
```
