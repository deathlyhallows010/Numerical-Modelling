```python
import numpy as np
import matplotlib.pyplot as plt
```

# Rungeâ€“Kutta method


```python
h = 0.1 # Discrtization 
```

```python
def RK4(x0, y0, x, h): 
    # N is the number of iteration to obtain the value of x form x0
    n = (int)((x - x0)/h)
    # initiating the value of y = y0
    y = y0 
    for i in range(1, n + 1):
        k1 = h * f(x0, y)
        k2 = h * f(x0 + 0.5 * h, y + 0.5 * k1)
        k3 = h * f(x0 + 0.5 * h, y + 0.5 * k2)
        k4 = h * f(x0 + h, y + k3)
  
        y = y + (1.0 / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4) 
                   
        x0 = x0 + h 
    return y

def euler( x0, y0, x, h ):
    n = (int)((x - x0)/h)
    y = y0 
    for i in range(1, n + 1):
        y = y + h * f(x0, y) 
        x0 = x0 + h 
    return y    
```

```python
def f(x, y): 
    return -y

x = np.arange(0,2,.1)
y = []
for i in x:
    y.append(RK4(0,1,i,h))

y_ = []
for i in x:
    y_.append(euler(0,1,i,h))
    
plt.plot(x,y,'o',c = 'b')
plt.plot(x,y_,'o',c = 'r')
plt.plot(x,np.exp(-x), c = 'k')
plt.legend(['RK4 Method','Eulers Method','Analytical Sol'])
plt.title('dy/dx = -y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
h = .05

def f(x, y): 
    return x**2 + y**2

x = np.arange(0,1,h)
y = []
for i in x:
    y.append(RK4(0,1,i,h))

y_ = []
for i in x:
    y_.append(euler(0,1,i,h))
    
print('The Function appraches to infinite at vaues of x at vicinity of 1 and so.')
plt.plot(x,y,'o',c = 'b')
plt.plot(x,y_,'o',c = 'r')
plt.legend(['RK4 Method','Eulers Method'])
plt.title('dy/dx = x^2 + y^2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def f(x, y): 
    return x*y

x = np.arange(0,2,.1)
y = []
for i in x:
    y.append(RK4(0,1,i,h))

y_ = []
for i in x:
    y_.append(euler(0,1,i,h))
    
plt.plot(x,y,'o',c = 'b')
plt.plot(x,y_,'o',c = 'r')
plt.plot(x,np.exp((x**2)/2), c = 'k')
plt.legend(['RK4 Method','Eulers Method','Analytical Sol'])
plt.title('dy/dx = xy')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def RK2(x0, y0, x, h): 
    # N is the number of iteration to obtain the value of x form x0
    n = (int)((x - x0)/h)
    # initiating the value of y = y0
    y = y0 
    for i in range(1, n + 1):
        k1 = h * f(x0, y)
        k2 = h * f(x0 + h, y + k1)
  
        y = y + (1/2)*(k1 + k2) 
                   
        x0 = x0 + h 
    return y

def Heuns_Method( x0, y0, x, h ):
    n = (int)((x - x0)/h)
    y = y0 
    for i in range(1, n + 1):
        t = y
        y = y + h * f(x0, y)
        y = t + (h/2)*(f(x0,t)+f(x0+h,y))
        x0 = x0 + h 
    return y   

def euler( x0, y0, x, h ):
    n = (int)((x - x0)/h)
    y = y0 
    for i in range(1, n + 1):
        y = y + h * f(x0, y) 
        x0 = x0 + h 
    return y
```

```python
def f(x, y): 
    return y - x

x = np.arange(0,200,1)
y = []
for i in x:
    y.append(RK2(0,0,i,h))

y_ = []
for i in x:
    y_.append(Heuns_Method(0,0,i,h))
    
plt.plot(x,y,'o',c = 'b')
plt.plot(x,y_,'o',c = 'r')
plt.plot(x,-np.exp(x)+x+1,'k')
plt.legend(['RK2 Method','Heuns Method','Analytical Sol'])
plt.title('dy/dx = y-x with intial condition x = 0 and y = 1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
