## Jacobi Method

```python
# Jacobi Method
# Solution of x in Ax=b using Jacobi Method
# Initailize 'A' 'b' & intial guess

A=np.array([[2, 1, 1],[-1, 3, 1],[1, 2, 4]])
b=np.transpose([4, -5, 6])
x=np.transpose([0, 0, 0])

n= len(x)
itr=20;

sol = np.zeros((itr+1,n))

for k in range(itr):
       
    for i in range(n):
        sigma=0        
        for j in range(n):
            if j!=i:
                sigma=sigma+A[i][j]*sol[k][j]
        
        sol[k+1][i]=(b[i]-sigma)/A[i][i]
        

    print("Solution: Iteration = ",k+1, "  ", sol[k+1])

```

```python
from numpy import array, zeros, diag, diagflat, dot

def jacobi(A,b,N,x=None):
    
    b_ = []
    
    if x is None:
        x = zeros(len(A[0]))                                                   
                                                                                                                                                                         
    D = diag(A)
    R = A - diagflat(D)
                                                                                                                                                                       
    for i in range(N):
        x = (b - dot(R,x)) / D
        b_.append(x)       
        
    return x, b_

A = array([[2, 1, 1],[-1, 3, 1],[1, 2, 4]])
b = array([4, -5, 6])
guess = array([0, 0, 0])

sol, b_ = jacobi(A,b,N=20,x=guess)
```

```python
for i in range(len(b_)):
    print("Iteration " ,i+1," ",b_[i])

print("A = ",A)
print("b = ",b)
print("Solution= ",sol)
```

```python
y_ = []
for i in range(20):
    y_.append(np.linalg.norm(b - np.dot(np.linalg.inv(A),b_[i])))

plt.plot(y_,'o')
plt.show()
```

## Gauss-Seidel

```python
# Jacobi Method
# Solution of x in Ax=b using Jacobi Method
# Initailize 'A' 'b' & intial guess

A=np.array([[2, 1, 1],[-1, 3, 1],[1, 2, 4]])
b=np.transpose([4, -5, 6])
x=np.transpose([0, 0, 0])

n= len(x)
itr=20;

sol = np.zeros((itr+1,n))

for k in range(itr):
       
    for i in range(n):
        sigma=0        
        for j in range(i):
            sigma=sigma+A[i][j]*sol[k+1][j]
        for j in range(i+1,n):
            sigma=sigma+A[i][j]*sol[k][j]
        
        sol[k+1][i]=(b[i]-sigma)/A[i][i]
        

    print("Solution: Iteration = ",k+1, "  ", sol[k+1])
```
