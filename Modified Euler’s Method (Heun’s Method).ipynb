```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler( x0, y0, x, h ):
    n = (int)((x - x0)/h)
    y = y0 
    for i in range(1, n + 1):
        y = y + h * f(x0, y) 
        x0 = x0 + h 
    return y

def Heuns_Method( x0, y0, x, h ):
    n = (int)((x - x0)/h)
    y = y0 
    for i in range(1, n + 1):
        t = y
        y = y + h * f(x0, y)
        y = t + (h/2)*(f(x0,t)+f(x0+h,y))
        x0 = x0 + h 
    return y   
```

```python
def f(x, y): 
    return -0.5*y

h = 0.1 # Discrtization 

x = np.arange(0,2,.1)

y = []
for i in x:
    y.append(euler(0,1,i,h))
    
plt.plot(x,y,'o',c = 'b')
plt.plot(x,np.exp(-0.5 * x), c = 'k')
plt.legend(['Eulers Method','Analytical Sol'])
plt.title('dy/dx = -0.5y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def f(x, y): 
    return y - x

x = np.arange(0,2,.1)
y = []
for i in x:
    y.append(euler(0,1,i,h))

y_ = []
for i in x:
    y_.append(Heuns_Method(0,1,i,h))
    
plt.plot(x,y,'o',c = 'b')
plt.plot(x,y_,'o',c = 'r')
plt.plot(x,x+1,'k')
plt.legend(['Eulers Method','Heuns Method','Analytical Sol'])
plt.title('dy/dx = y-x with intial condition x = 0 and y = 1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def f(x, y): 
    return y - x

x = np.arange(0,2,.1)
y = []
for i in x:
    y.append(euler(0,0,i,h))

y_ = []
for i in x:
    y_.append(Heuns_Method(0,0,i,h))
    
plt.plot(x,y,'o',c = 'b')
plt.plot(x,y_,'o',c = 'r')
plt.plot(x,-np.exp(x)+x+1,'k')
plt.legend(['Eulers Method','Heuns Method','Analytical Sol'])
plt.title('dy/dx = y-x with intial condition x = 0 and y = 0')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
